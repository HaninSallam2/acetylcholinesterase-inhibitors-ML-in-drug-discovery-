```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
target = new_client.target
target_query = target.search('Acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

```python
activity = new_client.activity
activity_dict = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
activity_df = pd.DataFrame.from_dict(activity_dict)
```

```python
activity_df
```

```python
activity_df = activity_df[activity_df.standard_value.notna()]
activity_df = activity_df[activity_df.canonical_smiles.notna()]

activity_df
```

```python
import pandas as pd
#transform pIC50
activity_df['standard_value'] = pd.to_numeric(activity_df['standard_value'], errors='coerce')
activity_df = activity_df.dropna(subset=['standard_value'])
activity_df.loc[activity_df['type'] == 'pIC50', 'standard_value'] = (
    10 ** -activity_df.loc[activity_df['type'] == 'pIC50', 'standard_value']
) * 10**6

activity_df.loc[activity_df['type'] == 'pIC50', 'type'] = 'IC50'
activity_df
```

```python
bioactivity_class = []
for i in activity_df.standard_value:
  if float(i) >= 300:
    bioactivity_class.append("inactive")
  elif float(i) <= 100:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
columns_to_keep = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
activity_df = activity_df[columns_to_keep]
activity_df['bioactivity_class'] = bioactivity_class
activity_df
```

```python
maxClm = activity_df['standard_value'].max()
print(maxClm)
minClm = activity_df['standard_value'].min()
print(minClm)
activity_df = activity_df[(activity_df['standard_value'] > 0) & (activity_df['standard_value'] <= 9999999)]
maxClm = activity_df['standard_value'].max()
print(maxClm)
minClm = activity_df['standard_value'].min()
print(minClm)

activity_df
```

```python
import numpy as np
def pIC50(df):
    df['standard_value'] = pd.to_numeric(df['standard_value'], errors='coerce')
    pIC50 = []
    for i in df['standard_value']:
        molar = i * (10**-6)  # Converts uM to M
        pIC50.append(-np.log10(molar))
    df['pIC50'] = pIC50
    x = df.drop('standard_value', axis=1)
    return x
activity_df_final = pIC50(activity_df)
activity_df_final
```

```python
# Take the average of pIC50 for duplicate canonical_smiles
activity_df_final = (
    activity_df_final
    .groupby('canonical_smiles', as_index=False)
    .agg({
        'molecule_chembl_id': 'first',  # Retain the first molecule_chembl_id (or another logic if needed)
        'pIC50': 'mean',
        'bioactivity_class': lambda x: x.mode()[0] if len(x.mode()) > 0 else x.iloc[0]  # Most common class

    })
)

# Check the resulting dataframe
activity_df_final
```

```python
activity_df_final.to_csv('bioactivity_preprocessed_Acetylcholinesterase_3classes_final.csv', index=False)
```

# **Oversampling**

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)

# Perform SMOTE to generate synthetic samples without undersampling
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new DataFrame with the resampled data
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['label'] = y_resampled

# Save the resampled data to CSV
df_resampled.to_csv('synthesized_data_from_original.csv', index=False)

# Check new class distribution
print(sorted(Counter(y_resampled).items()))
```
